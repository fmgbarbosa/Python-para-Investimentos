```
import yfinance as yf
import pandas as pd
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]

tickers = ' '.join(df['Symbol'].to_list())
data = yf.download(tickers,'2019-01-01','2019-12-31',group_by = 'ticker')
```

```
data_MSFT = data.MSFT['Adj Close']
data_APPL = data.AAPL['Adj Close']
```

```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
print(ax.lines)
line1 = ax.plot(data_MSFT,label='MSFT')
print(ax.lines)
line2 = ax.plot(data_APPL,label='AAPL')
print(ax.lines)

ax.legend()

print(line1)
print(line2)
```

```
print(ax.xaxis.label)
ax.set_xlabel('Data')

print(ax.yaxis.label)
ax.set_ylabel('Adj Close')

ax.set_title('Grafico Microsoft | Apple')

fig
```

```
fig = plt.figure()

gs = fig.add_gridspec(nrows=6,ncols=1)
ax1 = fig.add_subplot(gs[0:5,0])
ax2 = fig.add_subplot(gs[5,0])

ax1.plot(data.MSFT.Close)
#ax2.bar(data.MSFT.index,data.MSFT.Volume)
ax2.fill_between(data.MSFT.index,data.MSFT.Volume,0)

ax1.axes.get_xaxis().set_visible(False)
ax2.xaxis_date()
ax1.set_title('Fecho | Volume | MSFT')
```
